```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return (x+3)**2
```

```python
def df(x):
    return 2*(x+3)
```

```python
def gradient_descent(x_start,learning_rate,no_of_iter):
    x=x_start
    hist_x=[x]
    hist_y=[f(x)]
    for i in range(no_of_iter):
        grad=df(x)
        x=x-learning_rate*grad
        hist_x.append(x)
        hist_y.append(f(x))
    return x,hist_x,hist_y
```

```python
x=2
learning_rate=0.1
no_of_iter=50
min_x,hist_x,hist_y=gradient_descent(x,learning_rate,no_of_iter)
```

```python
x_label=np.linspace(-10,5,100)
y_label=f(x_label)

plt.plot(x_label,y_label,label="y=(x+3)^2",color="blue")
plt.scatter(hist_x,hist_y,color="red",label="Iterations",marker="x")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Gradient Descent for y=(x+3)^2")
plt.show()

print(f"Local Minimum found at {min_x : .4f}")
print(f"Corressponding y value for minimum is {f(min_x) : .2f}")
```
